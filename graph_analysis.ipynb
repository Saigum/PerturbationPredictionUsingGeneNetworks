```python
from gears import GEARS,PertData
import scanpy as sc
pert_data = PertData("vcc_custompert",default_pert_graph=False)
pert_data.load(data_path="vcc_custompert/arc")
pert_data.prepare_split(split = 'single', seed = 1)
pert_data.get_dataloader(batch_size = 32, test_batch_size = 128)
import numpy as np
## indices in adata.X where expression is nonzero for perturbed genesubset, grouped by gene.
non_zeros_gene_idx = {
    condition_name: np.flatnonzero(pert_data.adata.X[pert_data.adata.obs["condition_name"].values==condition_name].getnnz(axis=0))
    for condition_name in pert_data.adata.obs['condition_name'].unique()
}
pert_data.adata.uns["non_zeros_gene_idx"] = non_zeros_gene_idx
```

```python
gears_model = GEARS(pert_data, device = 'cpu', 
                        weight_bias_track = False, 
                        proj_name = 'pertnet', 
                        exp_name = 'pertnet')
gears_model.model_initialize(hidden_size = 64,)
gears_model.train(epochs = 4, lr = 1e-3)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
edge_indices = gears_model.config["G_go"].numpy().T
edge_list = list(map(tuple, edge_indices)) # Efficiently converts each row to a tuple
G = nx.Graph() 
G.add_edges_from(edge_list)

print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")
print(f"Graph density: {nx.density(G):.4f}")

# Degree distribution
degrees = [d for n, d in G.degree()]
print(f"Average degree: {np.mean(degrees):.2f}")
print(f"Max degree: {np.max(degrees)}")
print(f"Min degree: {np.min(degrees)}")

# Plot degree distribution (useful for large graphs)
degree_counts = nx.degree_histogram(G)
plt.figure(figsize=(10, 6))
plt.bar(range(len(degree_counts)), degree_counts, width=0.8, color='skyblue')
plt.title("Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("Number of Nodes")
plt.yscale('log') # Log scale is often useful for degree distributions
plt.show()


# Connected components
if nx.is_connected(G):
    print("Graph is connected.")
else:
    num_components = nx.number_connected_components(G)
    print(f"Number of connected components: {num_components}")
    # Get sizes of components
    component_sizes = [len(c) for c in nx.connected_components(G)]
    print(f"Largest component size: {np.max(component_sizes)}")
    print(f"Smallest component size: {np.min(component_sizes)}")
    print(f"Average component size: {np.mean(component_sizes):.2f}")

# Centrality measures (can be computationally intensive for very large graphs)
# Be cautious with these for 20,000+ nodes, start with smaller samples if needed.
# Betweenness centrality can be very slow.
degree_centrality = nx.degree_centrality(G)
print(f"Top 5 nodes by Degree Centrality: {sorted(degree_centrality.items(), key=lambda item: item[1], reverse=True)[:5]}")

closeness_centrality = nx.closeness_centrality(G)
print(f"Top 5 nodes by Closeness Centrality: {sorted(closeness_centrality.items(), key=lambda item: item[1], reverse=True)[:5]}")

eigenvector_centrality = nx.eigenvector_centrality(G, max_iter=1000) # May need more iterations
print(f"Top 5 nodes by Eigenvector Centrality: {sorted(eigenvector_centrality.items(), key=lambda item: item[1], reverse=True)[:5]}")

# PageRank (often faster than other advanced centralities for large graphs)
pagerank = nx.pagerank(G)
print(f"Top 5 nodes by PageRank: {sorted(pagerank.items(), key=lambda item: item[1], reverse=True)[:5]}")

# Community detection (requires additional libraries like `python-louvain` or `igraph`)
import community as co
partition = co.best_partition(G)
num_communities = len(set(partition.values()))
print(f"Number of communities (Louvain): {num_communities}")
```

```python
def get_genes_from_perts(perts):
    """
    Returns list of genes involved in a given perturbation list
    """

    if type(perts) is str:
        perts = [perts]
    gene_list = [p.split('+') for p in np.unique(perts)]
    gene_list = [item for sublist in gene_list for item in sublist]
    gene_list = [g for g in gene_list if g != 'ctrl']
    return list(np.unique(gene_list))

pert_list = get_genes_from_perts(pert_data.adata.obs.condition)
print(len(pert_list))
```
